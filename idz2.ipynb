```python
class parser:
    flag = True
    k = 0
    ch = ''
    s = ""
    result = ""
    res = "Good"
    n = 0
    mode = 1

    def go(self):
        while (self.mode == 1):
            self.outInfo()
            self.s = input()
            self.n = len(self.s)
            self.readNext()
            self.result = ""
            self.begin()

            if (self.k <= self.n):
                self.flag = False

            if (self.flag == False):
                self.res = "Error!"
            if (self.flag):
                print("Result: ")
                print(self.result)
            print(self.res)
            print("\n")

            self.res = "Good!"
            self.flag = True
            self.k = 0

    def readNext(self):
        self.result += self.ch
        if (self.k < self.n + 1):
            if (self.k < self.n):
                self.ch = self.s[self.k]
        else:
            self.flag = False

        self.k += 1


    def begin(self):
        if self.ch == '[':
            self.readNext()
            self.square()
        else:
            self.circle()

    def circle(self):
        if self.ch == '(':
            self.readNext()
            self.cont_circle()
            self.begin()



    def c2(self):
        if self.ch == '(':
            self.readNext()
            self.cont_circle()


    def cont_circle(self):
        if self.ch == ')':
            self.readNext()
            self.c2()
        else:
            self.c2()
            if self.ch == ')':
                self.readNext()
                self.c2()



    def square(self):
        if self.ch == ']':
            self.readNext()
            self.begin()
        else:
            self.begin()
            if self.ch == ']':
                self.readNext()
                self.begin()




    def outInfo(self):
        print("Правильные скобочные записи,\nу которых внутри квадратных могут быть любые, а внутри круглых - только круглые\n");
        print("Good! - запись верна, Error! - запись не верна")
        print("Пример правильной записи: (()())[()[[](())[]]]")
        print("Введите свою запись: ")


a = parser()
a.go()
```
